```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/olist_orders_dataset.csv", parse_dates=["order_purchase_timestamp",
                                                                  "order_approved_at",
                                                                  "order_delivered_carrier_date",
                                                                  "order_delivered_customer_date",
                                                                  "order_estimated_delivery_date"
                                                                 ]).dropna(how="all")
df.info()
df.head(3)
```

```python
delivered_df = df[df.order_delivered_customer_date.notna()]
delivered_df.info()
```

```python
#Zpoždění mezi očekávaným datem doručení a skutečným datem doručení
#kladné číslo - objednávaka dorazila dříve
#0 - objendávka dorazila v očekávané datum
#záporné číslo - objednávka dorazila později
delivered_df = delivered_df.assign(
    delay_days=(delivered_df.order_delivered_customer_date - delivered_df.order_estimated_delivery_date).dt.days)
delivered_df.head(3)
```

```python
delayed_df = delivered_df[delivered_df.delay_days > 0]
delivered_df.head(3)
```

```python
#Medián zpoždění
delayed_df.delay_days.median() #7
#Maximální zpoždění
delayed_df.delay_days.max() #188

#Kategorizace zpoždění
delayed_df = delayed_df.assign(
    delay_category=(
    np.where(delayed_df.delay_days <= 3, "1-3 days",
    np.where(delayed_df.delay_days <= 7, "4-7 days",
    np.where(delayed_df.delay_days <= 30, "8-30 days", 
             "30+ days"),
))))
delayed_df.head(10)
```

```python
#Histogram zpoždění 
counts, bins = np.histogram(delayed_df.delay_days, bins=[1, 4, 8, 30, np.inf])

plt.figure(figsize=(8,4))
x_labels = [
    f"{int(bins[i])}-{int(bins[i+1]) - 1}" if bins[i+1] != np.inf else f"{int(bins[i])}+"
    for i in range(len(counts))
]

plt.bar(x_labels, height=counts,width = 0.6)
plt.title("Delay Days Distribution")
plt.xlabel("Delay range (days)")
plt.ylabel("Number of orders")
plt.tight_layout()
plt.show()
```
