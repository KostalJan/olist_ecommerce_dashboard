```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/olist_orders_dataset.csv", parse_dates=["order_purchase_timestamp",
                                                                  "order_approved_at",
                                                                  "order_delivered_carrier_date",
                                                                  "order_delivered_customer_date",
                                                                  "order_estimated_delivery_date"
                                                                 ]).dropna(how="all")
df.info()
df.head(3)
```

```python
approval_df = df[df.order_approved_at.notna()].copy()
approval_df.info()
```

```python
#Doba schvÃ¡lenÃ­m objednÃ¡vky (v minutÃ¡ch a hodinÃ¡ch)
approval_df["approval_time_minutes"] = (
    ((approval_df.order_approved_at - approval_df.order_purchase_timestamp).dt.total_seconds() / 60).astype(int)
)
approval_df["approval_time_hours"] = (
    ((approval_df.order_approved_at - approval_df.order_purchase_timestamp).dt.total_seconds() / 3600).round(2)
)
approval_df
```

```python
approval_df.approval_time_minutes.describe()
```

```python
#ZÃ¡kladnÃ­ popis
approval_df.approval_time_minutes.mean().round() #624 min
approval_df.approval_time_minutes.median() #20 min
approval_df.approval_time_minutes.max() #27 0550 min = 187 days
```

```python
#Kategorie doby schvalovÃ¡nÃ­ objednÃ¡vek
approval_df["approval_category"] = (
    np.where(approval_df.approval_time_minutes <= 30, "0â€“30 min",
    np.where(approval_df.approval_time_minutes <= 120, "31â€“120 min",
    np.where(approval_df.approval_time_minutes <= 1440, "121â€“1440 min",
    "1440+ min"
            ))))
approval_df.head()
```

```python
#Df s absolutnÃ­m a relativnÃ­m pÅ™ehledem doby schvalovÃ¡nÃ­ objednÃ¡vek

categories = ["0â€“30 min", "31â€“120 min", "121â€“1440 min", "1440+ min"]

approval_category_counts = approval_df.approval_category.value_counts()
approval_category_pct = ((approval_category_counts / approval_category_counts.sum()) *100).round(1)

summary_df = pd.DataFrame(data={
    "Counts": approval_category_counts,
    "Percentage": approval_category_pct
}, index=categories)
summary_df
```

```python
#Histogram

bins = [0, 0.5, 2, 24, np.inf]
counts, bins = np.histogram(approval_df["approval_time_hours"], bins=bins)

x_labels = [
    "0â€“30 min",     # 0â€“0.5 h
    "31â€“120 min",   # 0.5â€“2 h
    "2â€“24 h",
    "24+ h"
]

plt.figure(figsize=(8, 4))
plt.bar(x_labels, counts, width=0.6)
plt.title("Order Approval Time Distribution")
plt.xlabel("Approval time")
plt.ylabel("Number of orders")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
